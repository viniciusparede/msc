# 1. Leitura dos dados

```python
repo_dir = os.path.dirname(os.getcwd())
data_dir = os.path.join(repo_dir, "data")

dataset = mmdatasdk.mmdataset(data_dir)
```

```python
words = dataset.computational_sequences["words"]
labels = dataset.computational_sequences["All Labels"]

display(words.keys())
display(labels.keys())
```

```python
random_key = random.choice(list(words.keys()))
display(random_key)
```

```python
word_test = words[random_key]
label_test = labels[random_key]

display(word_test.keys())
display(label_test.keys())
```

```python
word_test_features = word_test["features"][:]
word_test_intervals = word_test["intervals"][:]

label_test_features = label_test["features"][:]
label_test_intervals = label_test["intervals"][:]
```

```python
display(word_test_intervals)
print("\n")
display(word_test_features)
```

```python
display(label_test_intervals)
print("\n")
display(label_test_features)
```

# 2. Organizando o dataset

```python
word_test_features = np.array(
    [feature[0].decode("utf-8") for feature in word_test_features]
)
display(word_test_features)
```

```python
complete_sentences = []
current_sentence = []
current_label = None

for token_interval, token in zip(word_test_intervals, word_test_features):
    token_start, token_end = token_interval

    # Ignorar tokens marcados como "sp"
    if token == "sp":
        continue

    # Encontrar o rótulo correspondente
    for label_interval, label in zip(label_test_intervals, label_test_features):
        label_start, label_end = label_interval

        # Verificar se o token está dentro do intervalo do rótulo
        if label_start <= token_start < label_end:
            # Convertendo o rótulo para string (se necessário) para comparação
            label_str = str(label)
            if str(current_label) != label_str:
                # Se mudou o rótulo, começar uma nova frase
                if current_sentence:
                    complete_sentences.append(
                        (" ".join(current_sentence), current_label)
                    )
                    current_sentence = []
                current_label = label_str
            break

    current_sentence.append(token)

# Adicionar a última frase se ela existir
if current_sentence:
    complete_sentences.append((" ".join(current_sentence), current_label))

# Exemplo de saída
for sentence, label in complete_sentences:
    print(f"{sentence}")
    print("Sentiment, Happiness, Sadness, Anger, Fear, Disgust, Surprise")
    print(f"{label}\n")
```

```python
emotions = ["Sentiment", "Happiness", "Sadness", "Anger", "Fear", "Disgust", "Surprise"]
data = []
current_sentence = []
current_label = None

for token_interval, token in zip(word_test_intervals, word_test_features):
    token_start, token_end = token_interval

    if token == "sp":
        continue

    for label_interval, label in zip(label_test_intervals, label_test_features):
        label_start, label_end = label_interval

        if label_start <= token_start < label_end:
            # Comparar se as labels são diferentes
            if current_label is not None and not np.array_equal(current_label, label):
                # Adicionando a frase e o rótulo anterior ao DataFrame
                data.append([" ".join(current_sentence)] + list(current_label))
                current_sentence = []
            current_label = label
            break

    current_sentence.append(token)

# Adicionando a última frase e rótulo, se existirem
if current_sentence and current_label is not None:
    data.append([" ".join(current_sentence)] + list(current_label))

# Criar DataFrame
df = pd.DataFrame(data, columns=["Sentence"] + emotions)

# Exibir o DataFrame
display(df)
```
